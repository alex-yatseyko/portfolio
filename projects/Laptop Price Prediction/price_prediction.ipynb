## Preprocessing

```python
df = df.drop("Product", axis=1)
df
```

```python
df = df.join(pd.get_dummies(df.Company, dtype=int))
df = df.drop("Company", axis=1)
df
```

```python
df = df.join(pd.get_dummies(df.TypeName, dtype=int))
df = df.drop("TypeName", axis=1)
df
```

```python
df['ScreenResolution'] = df.ScreenResolution.str.split(' ').apply(lambda x: x[-1])
df
```

```python
df['Screen Width'] = df.ScreenResolution.str.split('x').apply(lambda x: x[0])
df['Screen Height'] = df.ScreenResolution.str.split('x').apply(lambda x: x[1])
df
```

```python
df = df.drop("ScreenResolution", axis=1)
df
```

```python
df['CPU_Brand'] = df.Cpu.str.split(' ').apply(lambda x: x[0])
df['CPU_Frequency'] = df.Cpu.str.split(' ').apply(lambda x: x[-1]).str[:3]
df = df.drop("Cpu", axis=1)
df["Ram"] = df["Ram"].str[:-2]
df
```

```python
df["Ram"] = df["Ram"].astype('int')

df['CPU_Frequency'] = df['CPU_Frequency'].replace({'3GH': '3.0', '2GH': '2.0'})
df["CPU_Frequency"] = df["CPU_Frequency"].astype('float')
df["Screen Width"] = df["Screen Width"].astype('int')
df["Screen Height"] = df["Screen Height"].astype('int')
df
```

```python
df.hist()
```

```python
df["Memory_Amount"] = df["Memory"].str.split(" ").apply(lambda x: x[0])
df["Memory_Type"] = df["Memory"].str.split(" ").apply(lambda x: x[1])
```

```python
df["Memory_Type"].value_counts()
```

```python
df["Memory_Amount"].value_counts()
```

```python
def turn_memory_info_MB(value):
    if "GB" in value:
        return float(value[:value.find("GB")]) * 1000
    elif "TB" in value:
        return float(value[:value.find("TB")]) * 1000000
```

```python
df["Memory_Amount"] = df["Memory_Amount"].apply(turn_memory_info_MB)
df = df.drop("Memory", axis=1)
df
```

```python
df = df.join(pd.get_dummies(df.Memory_Type))
df = df.drop("Memory_Type", axis=1)
```

```python
df["Weight"] = df["Weight"].str[:-2]
df["Weight"] = df["Weight"].astype('float')
```

```python
df["GPU_brand"] = df["Gpu"].str.split(" ").apply(lambda x: x[0])
df = df.drop("Gpu", axis=1)
```

```python
df = df.join(pd.get_dummies(df.OpSys))
df = df.drop("OpSys", axis=1)
```

```python
df
```

```python
# One hot encoding for CPU and GPU brand. The problem here is that we can get Intel for both so we need to rename cols somehow so it will not repeat
cpu_categories = pd.get_dummies(df["CPU_Brand"])
cpu_categories.columns = [col + "_CPU" for col in cpu_categories.columns]

df = df.join(cpu_categories)
df = df.drop("CPU_Brand", axis=1)
df
```

```python
df.columns
```

```python
# The same for GPU brand
gpu_categories = pd.get_dummies(df["GPU_brand"], dtype='int')
gpu_categories.columns = [col + "_GPU" for col in gpu_categories.columns]

df = df.join(gpu_categories)
df = df.drop("GPU_brand", axis=1)
df
```

```python
# Transforming all True/False endoded values to int
df = df*1
```

```python
df.columns
```

```python
df.corr(numeric_only=1)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(18, 15))
sns.heatmap(df.corr(numeric_only=1), annot=True, cmap="YlGnBu")
```

```python
target_correlations = df.corr(numeric_only=1)["Price_euros"].apply(abs).sort_values()
```

```python
target_correlations
```

```python
selected_features = target_correlations[-21:].index
selected_features = list(selected_features)
selected_features
```

```python
limited_df = df[selected_features]
limited_df
```

```python
plt.figure(figsize=(18, 15))
sns.heatmap(limited_df.corr(numeric_only=1), annot=True, cmap="YlGnBu")
```

## Model Development

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor

X, y = limited_df.drop("Price_euros", axis=1), limited_df["Price_euros"]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.15)

scaler = StandardScaler()

X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
forest = RandomForestRegressor()

forest.fit(X_train_scaled, y_train)
```

```python
forest.score(X_test_scaled, y_test)
```

```python
y_pred = forest.predict(X_test_scaled)
```

```python
plt.figure(figsize=(12,8))
plt.scatter(y_pred, y_test)
plt.plot(range(0, 5000), range(0, 5000), c="red")
```
