```python
def load_images_from_path(path, label):
    images = []
    labels = []

    for file in os.listdir(path):
        img = image.load_img(os.path.join(path, file), target_size=(224, 224, 3))
        images.append(image.img_to_array(img))
        labels.append((label))
        
    return images, labels

def show_images(images):
    fig, axes = plt.subplots(1, 8, figsize=(20, 20), subplot_kw={'xticks': [], 'yticks': []})

    for i, ax in enumerate(axes.flat):
        ax.imshow(images[i] / 255)

x_train = []
y_train = []
x_test = []
y_test = []
```

```python
images, labels = load_images_from_path('./data/ball', 0)
show_images(images)
    
x_train += images
y_train += labels
```

```python
images, labels = load_images_from_path('./data/cake', 1)
show_images(images)
    
x_train += images
y_train += labels
```

```python
images, labels = load_images_from_path('./data/guitar', 3)
show_images(images)
    
x_train += images
y_train += labels
```

```python
from keras.utils import to_categorical

x_train = np.array(x_train) / 255
x_test = np.array(x_test) / 255

y_train_encoded = to_categorical(y_train)
y_test_encoded = to_categorical(y_test)
```
