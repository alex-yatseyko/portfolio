```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url = "https://auto.ria.com/uk/"
```

```python
response = requests.get(url)
soup = BeautifulSoup(response.content, 'html.parser')
soup
```

```python
cars = soup.find('a', class_='link-showcase')
cars
```

```python
# Extract basic info about cars
cars_main = []
if cars:
    items = cars.find_all('div', class_='war_item')
    for item in items:
        value = item.find('div', class_='war_num').text.split('+')[0].strip('~')
        description = item.find('div', class_='war_desc').text.strip()
        cars_main.append({"Category": description, "Value": value})
```
